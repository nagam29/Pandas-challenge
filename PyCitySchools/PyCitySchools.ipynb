```python
# Dependencies and Setup
import pandas as pd
```

```python
# File to Load 
school_data_to_load ="/Users/mikan/Documents/Data Science Bootcamp/Pandas-challenge/PyCitySchools/schools_complete.csv"
student_data_to_load ="/Users/mikan/Documents/Data Science Bootcamp/Pandas-challenge/PyCitySchools/students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
school_data.head()
```

```python
student_data.head()
```

```python
# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_data_complete.head()
```

```python
######### District Summary ##########

# Calculate the total number of schools
# Select unique school names and save them in a list, schools
schools = school_data_complete["school_name"].unique()
print(schools)
# count the number of items in the list, schools
Total_Schools=len(schools)
print(Total_Schools)
```

```python
# Calculate the total number of students
# Select unique student IDs and save them in a list, students
students=school_data_complete["Student ID"].unique()
# count the number of items in the list, students
Total_Students=len(students)
print(Total_Students)
```

```python
# Calculate the total budget
# A specific budget for a specific school. So, data needs to be grouped by school
school_group = school_data_complete.groupby(["school_name"])
# One specific budget appears multiple times for a specific school.
# So, budget for min or max is actually the same for a specific scchool (only one budget for every school).
# Use min here to output one budget for each school
school_budget=school_group["budget"].min()
print(school_budget)
# Sum all the budgets
total_budget=sum(school_budget)
print(total_budget)
```

```python
# Calculate the average math score
avg_math_score=school_data_complete["math_score"].mean()
print(avg_math_score)
```

```python
# Calculate the average reading score
avg_reading_score=school_data_complete["reading_score"].mean()
print(avg_reading_score)
```

```python
# Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
passing_rate=(avg_math_score + avg_reading_score) / 2
print(passing_rate)
```

```python
# Calculate the percentage of students with a passing math score (70 or greater)
# Select rows with math_score >=70 and keep all the columns 
pass_math=school_data_complete.loc[school_data_complete["math_score"] >=70,:]
pass_math.head()
# Count student ID and calculate %
p_pass_math1=pass_math["Student ID"].count() / Total_Students * 100
print(p_pass_math1)
p_pass_math2=pass_math["Student ID"].count() / Total_Students
print(p_pass_math2)
```

```python
# Calculate the percentage of students with a passing reading score (70 or greater)
# Select rows with reading_score >=70 and keep all the columns 
pass_read=school_data_complete.loc[school_data_complete["reading_score"] >=70,:]
pass_read.head()
# Count student ID and calculate %
p_pass_read=pass_read["Student ID"].count() / Total_Students * 100
print(p_pass_read)
```

```python
####### School Summary ######

# One row for one school name. Groupby school name
s_summary = school_data_complete.groupby("school_name")

# Calculate total students per school
Total_Students = s_summary["student_name"].count()
Total_Students.head()
```

```python
# School Type
school_type=s_summary["type"].min()
school_type.head()
```

```python
# Calculate Total School Budget per school
Total_school_budget=s_summary["budget"].min()
Total_school_budget.head()
```

```python
# Calculate per student budget
Per_student_budget = Total_school_budget / Total_Students
Per_student_budget.head()
```

```python
# Calculate average math score
Avg_math_score = s_summary["math_score"].mean()
Avg_math_score.head()
```

```python
# Calculate average reading score
Avg_reading_score=s_summary["reading_score"].mean()
Avg_reading_score.head()
```

```python
# % Passing Math
# Students with passing scores were selected and stored in pass_math above
# Group by school name and count the number of students with the passing scores
pass_math.head()

pass_math2 = pass_math.groupby("school_name")
passed_math_stu=pass_math2["student_name"].count()
passed_math_stu.head()
# Calculate % passing math
p_passed_math = passed_math_stu / Total_Students * 100
p_passed_math.head()
```

```python
# % Passing Reading
# Students with passing scores were selected and stored in pass_math above
# Group by school name and count the number of students with the passing scores
pass_read.head()

pass_read2 = pass_read.groupby("school_name")
passed_read_stu=pass_read2["student_name"].count()
passed_read_stu.head()
# Calculate % passing reading
p_passed_read = passed_read_stu / Total_Students * 100
p_passed_read.head()
```

```python
# Overall Passing Rate (Average of the above two)
overall_pass_rate=(p_passed_read + p_passed_math)/2
overall_pass_rate.head()
```

```python
# Create a dataframe to hold the above results
School_summary = pd.DataFrame({"School Type" : school_type,
                "Total Students" : Total_Students,
                "Total School Budget" : Total_school_budget.map('${0:,.2f}'.format),
                "Per Student Budget" : Per_student_budget.map('${0:,.2f}'.format),
                "Average Math Score" : Avg_math_score,
                "Average Reading Score" : Avg_reading_score,
                "% passing Math" : p_passed_math,        
                "% Passing Reading" : p_passed_read,
                "Overall Passing Rate" : overall_pass_rate
                })

School_summary.index.name=None
School_summary.head()
```

```python
######## Top Performing Schools (By Passing Rate) #######
Desc_school_summary = School_summary.sort_values(by='Overall Passing Rate', ascending=False)
Desc_school_summary.index.name=None 
Desc_school_summary.head()
```

```python
######### Bottom Performing Schools (By Passing Rate) ##########
Asce_school_summary = School_summary.sort_values(by='Overall Passing Rate', ascending=True)
Asce_school_summary.index.name=None 
Asce_school_summary.head()
```

```python
##### Math Scores by Grade #####

# Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.
# Create a pandas series for each grade. Hint: use a conditional statement.
# Use the syntax like the one below from the class:
#good_movies_df = movie_file_df.loc[movie_file_df["IMDB"] > 7, [
  #  "FILM", "IMDB", "IMDB_user_vote_count"]]
grade_9 = school_data_complete.loc[school_data_complete["grade"]=="9th", ["school_name", "math_score"]]
grade_9.head()
grade_10 = school_data_complete.loc[school_data_complete["grade"]=="10th", ["school_name", "math_score"]]
grade_11 = school_data_complete.loc[school_data_complete["grade"]=="11th", ["school_name", "math_score"]]
grade_12 = school_data_complete.loc[school_data_complete["grade"]=="12th", ["school_name", "math_score"]]


# Group each series by school
grouped_grade_9=grade_9.groupby("school_name")
grouped_grade_10=grade_10.groupby("school_name")
grouped_grade_11=grade_11.groupby("school_name")
grouped_grade_12=grade_12.groupby("school_name")

# Calculate average math score per school
Avg_math_9 = grouped_grade_9["math_score"].mean()
Avg_math_10 = grouped_grade_10["math_score"].mean()
Avg_math_11 = grouped_grade_11["math_score"].mean()
Avg_math_12 = grouped_grade_12["math_score"].mean()
Avg_math_12
```

```python
# Merge 4 dataframes using an inner join
# First, merge 9th grade and 10th grade on school names
merge9_10_df = pd.merge(Avg_math_9, Avg_math_10,  on="school_name")

# Merge 11th and 12th grades on school names
merge11_12_df = pd.merge(Avg_math_11, Avg_math_12,  on="school_name")

# Merge merge9_10_df and merge11_12_df
merge_all_df = pd.merge(merge9_10_df, merge11_12_df, on="school_name")

# Rename the headers
merge_all2_df=merge_all_df.rename(columns={"math_score_x_x": "9th", "math_score_y_x": "10th",
                            "math_score_x_y":"11th", "math_score_y_y": "12th"})
merge_all2_df.index.name=None 
merge_all2_df
```

```python
######### Reading Score by Grade ##########

# Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.
# Create a pandas series for each grade. Hint: use a conditional statement.
# Use the syntax like the one below from the class:
#good_movies_df = movie_file_df.loc[movie_file_df["IMDB"] > 7, [
  #  "FILM", "IMDB", "IMDB_user_vote_count"]]
grade_9r = school_data_complete.loc[school_data_complete["grade"]=="9th", ["school_name", "reading_score"]]
grade_9r.head()
grade_10r = school_data_complete.loc[school_data_complete["grade"]=="10th", ["school_name", "reading_score"]]
grade_11r = school_data_complete.loc[school_data_complete["grade"]=="11th", ["school_name", "reading_score"]]
grade_12r = school_data_complete.loc[school_data_complete["grade"]=="12th", ["school_name", "reading_score"]]


# Group each series by school
grouped_grade_9r=grade_9r.groupby("school_name")
grouped_grade_10r=grade_10r.groupby("school_name")
grouped_grade_11r=grade_11r.groupby("school_name")
grouped_grade_12r=grade_12r.groupby("school_name")

# Calculate average math score per school
Avg_reading_9 = grouped_grade_9r["reading_score"].mean()
Avg_reading_10 = grouped_grade_10r["reading_score"].mean()
Avg_reading_11 = grouped_grade_11r["reading_score"].mean()
Avg_reading_12 = grouped_grade_12r["reading_score"].mean()
Avg_reading_12
```

```python
# Merge 4 dataframes using an inner join
# First, merge 9th grade and 10th grade on school names
merge9r_10r_df = pd.merge(Avg_reading_9, Avg_reading_10,  on="school_name")

# Merge 11th and 12th grades on school names
merge11r_12r_df = pd.merge(Avg_reading_11, Avg_reading_12,  on="school_name")

# Merge merge9_10_df and merge11_12_df
merge_all_reading_df = pd.merge(merge9r_10r_df, merge11r_12r_df, on="school_name")
merge_all_reading_df

# Rename the headers
merge_all2_reading_df=merge_all_reading_df.rename(columns={"reading_score_x_x": "9th", "reading_score_y_x": "10th",
                           "reading_score_x_y":"11th", "reading_score_y_y": "12th"})
merge_all2_reading_df.index.name=None 
merge_all2_reading_df
```

```python
###### Scores by School Spending #######

# Use School_summary data frame since this is grouped by school names and has
# School Type, Total Students, Total School Budget, Per Student Budget, Average Math Score, Average Reading Score, 
# % passing Math, % Passing Reading, Overall Passing Rate

# Currently Per Student Budget is string since I added $
# just create another data frame without $ format
School_summary2 = pd.DataFrame({"School Type" : school_type,
                "Total Students" : Total_Students,
                "Total School Budget" : Total_school_budget.map('${0:,.2f}'.format),
                "Per Student Budget" : Per_student_budget,
                "Average Math Score" : Avg_math_score,
                "Average Reading Score" : Avg_reading_score,
                "% passing Math" : p_passed_math,        
                "% Passing Reading" : p_passed_read,
                "Overall Passing Rate" : overall_pass_rate
                })


School_summary2.head()
```

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

# create a new variable to store binned Spending Ranges (Per Student)
School_summary2["Spending Ranges (Per Student)"] = pd.cut(School_summary2["Per Student Budget"], spending_bins, labels=group_names)
School_summary2.head()
```

```python
# Group by Spending Ranges (Per Student)
# Keep "Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "Overall Passing Rate"

School_summary2_g = School_summary2.groupby(["Spending Ranges (Per Student)"])
School_summary2_g = School_summary2_g["Average Math Score", "Average Reading Score","% passing Math", "% Passing Reading", "Overall Passing Rate"].mean()
School_summary2_g
```
